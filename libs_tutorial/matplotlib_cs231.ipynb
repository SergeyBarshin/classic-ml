# Matplotlib
Matplotlib — это библиотека для построения графиков в Python. Ее модуль matplotlib.pyplot предоставляет интерфейс, схожий с MATLAB, что делает его удобным для быстрого создания графиков.

### Plotting
Основная функция `plot` позволяет строить 2D графики.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 3 * np.pi, 0.1) # от нуля до 3pi, с шагом в 0.1
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.plot(x, y_sin)
plt.plot(x, y_cos)
plt.xlabel('x axis label')
plt.ylabel('y axis label')
plt.title('Sine and Cosine')
plt.legend(["sin", "cos"])
plt.show()

```

### Subplots

```python
x = np.arange(0, 3 * np.pi, 0.1) # от нуля до 3pi, с шагом в 0.1
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.subplot(2, 1, 1)
plt.plot(x, y_sin, label='sin(x)', color='b')
plt.title("Sin")

plt.subplot(2, 1, 2)
plt.plot(x, y_cos, label='cos(x)', color='g', linestyle='-.')
plt.title("Cos")

plt.show()
```
